```python
###This is to generate the sinusoidal wave form which is to be fed to the FIR Low Pass filter. As you can see a waveform is generated which
###composes of frequencies 200kHz and 500kHz. This waveform is sampled and the values are thenutilized as the input to the filter



import numpy as np
import matplotlib.pyplot as plt

# Define parameters
frequency1 = 200e3  # first signal frequency in Hz (200 kHz)
frequency2 = 500e3  # second signal frequency in Hz (500 kHz)
sampling_rate = 1e6  # sampling rate in Hz (1 MHz)
duration = 1e-3  # signal duration in seconds

# Calculate the number of samples
num_samples = int(duration * sampling_rate)

# Generate time vector and mixed signal
t = np.linspace(0, duration, num_samples, endpoint=False)
signal = np.sin(2 * np.pi * frequency1 * t) + np.sin(2 * np.pi * frequency2 * t)

# Display first 10 samples with comma-separated values
print("First 1000 samples:", ', '.join(f"{value:.6f}" for value in signal[:1000]))

# Plotting the mixed signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal, label='Signal: sin(2π * 200k * t) + sin(2π * 500k * t)', color='blue')
plt.title('Time vs Samples for Mixed Signal (200 kHz + 500 kHz)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.xlim(0, 0.0001)  # Limit x-axis to the first 0.1 ms for better visualization
plt.ylim(-2, 2)      # Set y-axis limits
plt.legend()
plt.show()
```
